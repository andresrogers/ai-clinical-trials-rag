# 03 - Retrieval Strategies & Ground Truth Construction

This notebook focuses on two professional RAG components:
1. **Ground Truth Pipeline (Defensive)**:
   - Retrieve top-5 raw chunks per question.
   - Draft 1 (extractive) with citations.
   - Human/expert review.
   - Draft 2 refinement (no inference).
   - Final JSON ground truth with metadata.
2. **Hybrid Retrieval**: Dense + BM25 for question-time evaluation.
3. **Reranking**: Cross-encoder to boost the most relevant chunks.

```python
import json

# Step 1: Retrieve top-k raw chunks per question
retrieved_records = []
for idx, row in df.iterrows():
    question_raw = str(row["question"]).strip()
    if not question_raw:
        continue
    nct_id = row["nct_id"] if "nct_id" in df.columns and pd.notna(row["nct_id"]) else None
    study_title = row["brief_title"] if "brief_title" in df.columns and pd.notna(row["brief_title"]) else None

    # Keep a verbose question for logging/output, but retrieve using the raw question
    question = f"For Clinical Study {nct_id} titled '{study_title}': " + question_raw

    results = retrieve_chunks(
        collection=collection,
        embedder=embedder,
        question=question_raw,
        nct_id=nct_id,
        top_k=TOP_K,
        collection_dim=collection_dim,
    )
    chunks = results.get("documents", [[]])
    metadatas = results.get("metadatas", [[]])
    ids = results.get("ids", [[]])

    retrieved_records.append(
        {
            "row_id": int(idx),
            "nct_id": nct_id,
            "question": question,
            "retrieved_chunks": chunks[0] if chunks else [],
            "metadatas": metadatas[0] if metadatas else [],
            "chunk_ids": ids[0] if ids else [],
        }
    )

with open(RETRIEVED_CONTEXTS_PATH, "w", encoding="utf-8") as f:
    json.dump(retrieved_records, f, ensure_ascii=False, indent=2)

print(f"Saved {len(retrieved_records)} retrieved context entries to {RETRIEVED_CONTEXTS_PATH}")
retrieved_records[:2]
```

```python
import json

# Step 2: LLM Draft 1 (extractive, citation-focused)
with open(RETRIEVED_CONTEXTS_PATH, "r", encoding="utf-8") as f:
    retrieved_records = json.load(f)

# sample_size = min(10, len(retrieved_records))  # 10-row sample for testing
sample_size = len(retrieved_records)             # full run
sample_records = retrieved_records[:sample_size]
total = len(sample_records)

draft1_records = []
for i, rec in enumerate(sample_records, start=1):
    chunks = rec.get("retrieved_chunks", [])
    if not chunks:
        logger.warning(
            f"[Draft1] {i}/{total} row_id={rec.get('row_id')} status=no_chunks"
        )
        draft1_records.append(
            {
                **rec,
                "draft_answer": "N/A",
                "draft_citations": [],
                "draft_confidence": "low",
                "draft_missing_info": "No retrieved contexts available.",
            }
        )
        continue

    retrieved_chunks_text = format_retrieved_chunks(chunks)

    try:
        parsed = invoke_chain_with_retries(
            draft_chain_raw,
            {
                "retrieved_chunks": retrieved_chunks_text,
                "question": rec["question"],
                "format_instructions": draft_parser.get_format_instructions(),
            },
            max_retries=2,
            min_citations=1,
            allow_empty_answer=False,
            sleep_seconds=0.5,
        )

        answer = parsed.get("answer")
        citations = parsed.get("citations") or []
        status = "ok"
        if not isinstance(answer, str) or not answer.strip() or answer.strip().upper() == "N/A":
            status = "empty_answer"
        elif len(citations) == 0:
            status = "no_citations"

        logger.info(
            f"[Draft1] {i}/{total} row_id={rec.get('row_id')} status={status} "
            f"citations={len(citations)}"
        )

        draft1_records.append(
            {
                **rec,
                "draft_answer": parsed.get("answer"),
                "draft_citations": parsed.get("citations"),
                "draft_confidence": parsed.get("confidence"),
                "draft_missing_info": parsed.get("missing_info"),
            }
        )
    except Exception as e:
        logger.error(
            f"[Draft1] {i}/{total} row_id={rec.get('row_id')} status=error error={e}"
        )
        draft1_records.append(
            {
                **rec,
                "draft_answer": "N/A",
                "draft_citations": [],
                "draft_confidence": "low",
                "draft_missing_info": f"LLM error: {e}",
            }
        )

with open(DRAFT1_PATH, "w", encoding="utf-8") as f:
    json.dump(draft1_records, f, ensure_ascii=False, indent=2)

print(f"Saved Draft 1 answers to {DRAFT1_PATH} (sample_size={sample_size})")
draft1_records[:2]
```

## Expert Review Instructions (to generate Draft 2)

Before running Draft 2, update `expert_review.csv` with your expert corrections:

- **`expert_notes`**: Add brief rationale or evidence notes based on the retrieved chunks.
- **`revised_answer`**: Provide your corrected answer (verbatim). Leave blank to accept the Draft 1 answer.

Guidelines:
- Keep edits grounded in the retrieved chunks (no external knowledge).
- If the correct response is negative, explicitly say **“not mentioned in these contexts.”**
- If there is no supporting evidence, leave `revised_answer` empty and explain in `expert_notes`; the model will output **`answer = "N/A"`** with **low confidence**.

```python
# Step 4: LLM Draft 2 (refinement using expert-reviewed drafts)
# If revising the documents manually, skip this step

import ast
import json
import re
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from biotech_rag.retrieval.context_retrieval import (
    format_retrieved_chunks,
    retrieve_chunks,
 )

# Mock expert review using LLM + retrieval (clarify -> retrieve -> answer -> revise)
review_df = pd.read_csv(REVIEW_PATH)
# Ensure expert_notes and revised_answer are string dtype to avoid pandas dtype warnings
for _col in ("expert_notes", "revised_answer"):
    if _col not in review_df.columns:
        review_df[_col] = ""
    else:
        review_df[_col] = review_df[_col].fillna("").astype(str)

def _parse_citations(value: object) -> list[str]:
    if value is None or (isinstance(value, float) and pd.isna(value)):
        return []
    if isinstance(value, list):
        return [str(v) for v in value]
    if isinstance(value, str):
        text = value.strip()
        if not text:
            return []
        try:
            parsed = json.loads(text)
            if isinstance(parsed, list):
                return [str(v) for v in parsed]
        except Exception:
            pass
        try:
            parsed = ast.literal_eval(text)
            if isinstance(parsed, list):
                return [str(v) for v in parsed]
        except Exception:
            pass
        return [text]
    return [str(value)]

def _skip_missing_info(text: object) -> bool:
    if text is None:
        return True
    normalized = re.sub(r"\s+", " ", str(text)).strip().lower()
    if normalized in {"", "none", "no missing information", "no missing info", "n/a", "na"}:
        return True
    if "no missing information" in normalized or "no missing info" in normalized:
        return True
    return False

clarify_prompt = ChatPromptTemplate.from_template(
    "Role: Clinical trial expert reviewer.\n\n"
    "Question: {question}\n"
    "Missing info: {draft_missing_info}\n\n"
    "Task:\n"
    "Write ONE clarifying question that would resolve the missing info.\n"
    "Return ONLY the question text. No markdown."
 )
clarify_chain = clarify_prompt | llm | StrOutputParser()

answer_prompt = ChatPromptTemplate.from_template(
    "Role: Clinical trial expert reviewer.\n\n"
    "Clarifying question: {clarifying_question}\n"
    "NCT ID: {nct_id}\n"
    "Retrieved chunks:\n{retrieved_chunks}\n\n"
    "Task:\n"
    "Answer the clarifying question using ONLY the retrieved chunks.\n"
    "If not found, respond: \"Not mentioned in these contexts.\"\n"
    "Return only the answer text. No markdown."
 )
answer_chain = answer_prompt | llm | StrOutputParser()

revise_prompt = ChatPromptTemplate.from_template(
    "Role: Clinical trial expert reviewer.\n\n"
    "Question: {question}\n"
    "NCT ID: {nct_id}\n"
    "Draft answer: {draft_answer}\n"
    "Expert notes (answer to clarifying question): {expert_notes}\n\n"
    "Task:\n"
    "Improve the draft answer using ONLY the expert notes.\n"
    "If evidence is missing, answer \"N/A\" and say \"not mentioned in these contexts.\"\n"
    "Return ONLY the revised answer text. No JSON, no markdown."
 )
revise_chain = revise_prompt | llm | StrOutputParser()

total = len(review_df)
updated = 0
for i, row in review_df.iterrows():
    missing_info = row.get("draft_missing_info", "")
    if _skip_missing_info(missing_info):
        continue

    try:
        clarifying_question = clarify_chain.invoke(
            {
                "question": row.get("question", ""),
                "draft_missing_info": missing_info,
            }
        )
        clarifying_question = re.sub(r"\s+", " ", str(clarifying_question)).strip()
        if not clarifying_question:
            clarifying_question = str(missing_info).strip()

        nct_id = row.get("nct_id", "")
        if pd.isna(nct_id):
            nct_id = ""
        nct_id_str = str(nct_id).strip()
        search_query = f"{nct_id_str} {clarifying_question}".strip()

        results = retrieve_chunks(
            collection=collection,
            embedder=embedder,
            question=search_query,
            nct_id=nct_id_str if nct_id_str else None,
            top_k=TOP_K,
            collection_dim=collection_dim,
        )
        chunks = results.get("documents", [[]])
        chunks_list = chunks[0] if chunks else []
        retrieved_chunks_text = format_retrieved_chunks(chunks_list)
        if not retrieved_chunks_text:
            retrieved_chunks_text = ""

        existing = _parse_citations(row.get("draft_citations", []))
        if chunks_list:
            start_idx = len(existing) + 1
            appended = [
                f"Chunk {start_idx + j}: {chunk}"
                for j, chunk in enumerate(chunks_list)
                if str(chunk).strip()
            ]
            review_df.at[i, "draft_citations"] = existing + appended

        if not retrieved_chunks_text:
            expert_notes = "Not mentioned in these contexts."
        else:
            expert_notes = answer_chain.invoke(
                {
                    "clarifying_question": clarifying_question,
                    "nct_id": nct_id_str,
                    "retrieved_chunks": retrieved_chunks_text,
                }
            )
            expert_notes = str(expert_notes).strip()
            if not expert_notes:
                expert_notes = "Not mentioned in these contexts."

        revised_answer = revise_chain.invoke(
            {
                "question": row.get("question", ""),
                "nct_id": nct_id_str,
                "draft_answer": row.get("draft_answer", ""),
                "expert_notes": expert_notes,
            }
        )
        revised_answer = str(revised_answer).strip()

        review_df.at[i, "expert_notes"] = expert_notes
        review_df.at[i, "revised_answer"] = revised_answer
        updated += 1
        logger.info(
            f"[Expert-LLM] {i + 1}/{total} row_id={row.get('row_id')} status=ok"
        )
    except Exception as e:
        logger.warning(
            f"[Expert-LLM] {i + 1}/{total} row_id={row.get('row_id')} status=error error={e}"
        )
        continue

review_df.to_csv(REVIEW_PATH, index=False)
print(f"Saved expert review (LLM) to {REVIEW_PATH} | updated rows: {updated}")
review_df.head(2)
```

## Part 2: Advanced Hybrid Retrieval & Reranking

Now that we have a target (Ground Truth), we will implement a professional hybrid retrieval pipeline. 
Clinical data requires high precision (finding the exact drug name) and high recall (understanding the medical context).

We will compare:
1. **Vector Only** (Standard Semantic Search)
2. **Hybrid** (Vector + BM25)
3. **Hybrid + Reranking** (Using a Cross-Encoder)

```python
from biotech_rag.retrieval.retrievers import create_hybrid_retriever, retrieve_with_rerank
from biotech_rag.indexing.vectorstore import get_chroma_class
from biotech_rag.indexing.embedders import Embedder
from langchain_core.embeddings import Embeddings

# Professional Adapter for our Embedder to work with LangChain
class LangChainEmbedderAdapter(Embeddings):
    def __init__(self, embedder):
        self.embedder = embedder
    def embed_documents(self, texts):
        return self.embedder.embed(texts)
    def embed_query(self, text):
        return self.embedder.embed([text])[0]

# Wrap our existing embedder
lc_embeddings = LangChainEmbedderAdapter(embedder)

# Load full vectorstore into LangChain wrapper
ChromaClass = get_chroma_class()
vstore = ChromaClass(
    persist_directory=str(VSTORE_DIR), 
    embedding_function=lc_embeddings,
    collection_name="clinical_trials",
)

# Fetch all docs from Chroma for BM25 indexing (required for sparse search)
all_data = collection.get()
all_texts = all_data['documents']
all_metadatas = all_data['metadatas']

# Initialize Hybrid Retriever
hybrid_retriever = create_hybrid_retriever(
    vectorstore=vstore,
    all_texts=all_texts,
    all_metadatas=all_metadatas,
    k=10,
)
```

```python
# Comparison Test
sample_query = "What are the inclusion criteria for the NCT04205812 clinical trial?"

print("--- VECTOR ONLY (Top 5) ---")
res_v = vstore.similarity_search(sample_query, k=5)
for d in res_v:
    print(f"[{d.metadata.get('nct_id')}] {d.page_content[:150]}...")

print("\n--- HYBRID (Vector+BM25) (Top 5) ---")
if hasattr(hybrid_retriever, "invoke"):
    hybrid_docs = hybrid_retriever.invoke(sample_query)
else:
    try:
        hybrid_docs = hybrid_retriever.get_relevant_documents(sample_query)
    except Exception:
        hybrid_docs = hybrid_retriever(sample_query) if callable(hybrid_retriever) else []

if not hybrid_docs:
    print("No hybrid results.")
else:
    for d in hybrid_docs[:5]:
        print(f"[{d.metadata.get('nct_id')}] {d.page_content[:150]}...")

print("\n--- HYBRID + RERANK (Top 5) ---")
res_rerank = retrieve_with_rerank(hybrid_retriever, sample_query, top_n=5)
for d in res_rerank:
    score = d.metadata.get('rerank_score', 0.0)
    print(f"[{d.metadata.get('nct_id')}] (Score: {score:.2f}) {d.page_content[:150]}...")
```

```python
# Run retrieval evaluation (sample of 10 to start)
from biotech_rag.evaluation.retrieval_eval import evaluate_retrieval

# Use a small random sample first to avoid long runs and NCT-id bias
result = evaluate_retrieval(
    vstore=vstore,
    hybrid_retriever=hybrid_retriever,
    rerank_fn=retrieve_with_rerank,
    final_gt_path=FINAL_GROUND_TRUTH_PATH,
    top_k=5,
    sample_size=10,
    sample_strategy="random",
    seed=42,
    filter_by_nct_id=False, # False gives excellent, True gives mediocre
    rerank_candidate_k=5,
    verbose=False,
 )
print('Evaluation summary:')
print('\nDetails saved at:', result['details_path'])
result['summary']
```
